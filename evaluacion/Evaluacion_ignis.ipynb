#**Evaluación Del Modelo Propuesto IGNIS** 

### *Librerías Necesarias*

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
```

### *Se leen las dos Lista a comparar (Real y Predicha)*

```python
y_true=pd.read_csv("list_test.csv",sep=",")
y_pred3=pd.read_csv("list_predict_ignis.csv",sep=",")
```

### *Se genera la Matriz de Confusión*

```python
cm3=confusion_matrix(y_true,y_pred3)
print(cm3)
```

```python
col=["0","1"]
df_cm3=pd.DataFrame(cm3, index=col, columns=col)
grafica=sns.heatmap(cm3, annot=True)
grafica.set(xlabel="Predicciones", ylabel="Verdaderos", title="IGNIS")
plt.show()
```

### *Se calculan las Métricas de Evaluación*

```python
metricas3=classification_report(y_true, y_pred3)
print(metricas3)
```
